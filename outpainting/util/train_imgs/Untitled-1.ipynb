```python
import cv2
import os

def resize_images_in_folder(input_folder, output_folder, target_size=(256, 256)):
    if not os.path.exists(output_folder):
        os.makedirs(output_folder)

    for filename in os.listdir(input_folder):
        if filename.endswith(('.jpg', '.jpeg', '.png', '.gif')):  # 可以根据需要添加更多的图片格式
            input_path = os.path.join(input_folder, filename)
            output_path = os.path.join(output_folder, filename)

            # 读取图像文件
            img = cv2.imread(input_path)
            # 调整图像尺寸
            resized_img = cv2.resize(img, target_size, interpolation=cv2.INTER_AREA)  # 使用INTER_AREA插值方法
            # 保存调整尺寸后的图像
            cv2.imwrite(output_path, resized_img)
            print(f"Resized {filename} to {target_size}")

# 指定输入和输出文件夹的路径
input_folder = "/home/jonathanliu/Desktop/hgsh/outpainting_srn/train_imgs"
output_folder = "/home/jonathanliu/Desktop/hgsh/outpainting_srn/train_imgs"

# 调用函数将输入文件夹中的图片转换为256x256的尺寸并保存到输出文件夹
resize_images_in_folder(input_folder, output_folder)
```
